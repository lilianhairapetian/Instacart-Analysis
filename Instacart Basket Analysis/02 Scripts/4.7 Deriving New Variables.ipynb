## Creating a subset dataframe to include only the first 1mln rows 

```python
df = df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
# creating a new funtion namely price_label which will use if-else conditions to labed the data according to conditions
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# applying or using the function created above where df['price_range'] would create a new column, where the labels will be placed
# axis = 1 stands for rows
# axis = 0 would stand for columns 
df['price_range'] = df.apply(price_label, axis=1) 
```

```python
df['price_range'].value_counts (dropna = False)
```

```python
df['prices'].max()
```

### using loc() function to performe and separate price ranges same as above

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```
